1. Ayer creamos un par de gráficas para los tickets de la categoría "Software" y "Hardware", donde visualizamos el número de clientes que teníamos en cada grupo de satisfacción. Hoy tendréis que hacer dos gráficas con subplots:
    1.1. En la primera debéis mostrar la cantidad de tickets en función del tipo (si es Problema o Solicitud), cada subplot corresponderá a cada una de las categorías. Debéis:
* Poner nombre a los ejes de cada uno de los subplots
* Poner un títulos a cada uno de los subplots
* Poner el color de las líneas en negro.
* Establecer el tamaño de la gráfica a 15 x 5
* Ponerle marcadores a las gráficas
* No incluir leyenda

```python
df = pd.read_csv("../Pandas/data/relacion_años_pair5.csv")
df.head()
```

```python
df_hardware = df[df["categoría"] == "Hardware"]
df_hardware.head()
```

```python
df_software = df[df["categoría"] == "Software"]
df_software.head()
```

```python
df_hardware_tipo = pd.DataFrame(df_hardware["tipo"].value_counts()).reset_index()
df_hardware_tipo.head()
```

```python
df_software_tipo = pd.DataFrame(df_software["tipo"].value_counts()).reset_index()
df_software_tipo.head()
```

```python
fig,ax = plt.subplots(1,2, figsize = (15,5))

ax[0].plot(df_hardware_tipo["tipo"], df_hardware_tipo["count"], color = "black", marker = "^")
ax[1].plot(df_software_tipo["tipo"], df_software_tipo["count"], color = "black", marker = "^")

ax[0].set_title("Gráfico de Hardware")
ax[1].set_title("Gráfico de Software")

ax[0].set_xlabel("Tipo de tickets")
ax[0].set_ylabel("Cantidad de tickets")

ax[1].set_xlabel("Tipo de tickets")
ax[1].set_ylabel("Cantidad de tickets")

plt.savefig('graficas/graficas_lineal_tickets_soft_hard.jpg')
plt.show()
```

```python
df_software_sat = pd.DataFrame(df_software["satisfacción"].value_counts()).reset_index()
df_software_sat.head()
```

2. Cread un subplot con dos gráficas, dónde en la primera tengamos el número de empleados por departamento y en la otra por tipo de puesto. Debéis:
   * Poner leyenda a la gráfica
   * Cada línea debe ir en un color
   * Ponerle nombre a los ejes y a la gráfica.
   * Poner marcadores a las lineas.

```python
df2 = pd.read_csv("../Pandas/data/relacion_empleados_puestos_pair5.csv")
df2.head()
```

```python
df_departamento= pd.DataFrame(df2["departamento"].value_counts()).reset_index()
df_departamento.head()
```

```python
df_puesto= pd.DataFrame(df2["tipo_puesto"].value_counts()).reset_index()
df_puesto
```

```python
fig, ax = plt.subplots(2,1, figsize = (20,30))

ax[0].plot(df_departamento["departamento"], df_departamento["count"], marker='o', color='b', label='Empleados por Departamento')
ax[0].legend()

ax[1].plot(df_puesto['tipo_puesto'], df_puesto['count'], marker='o', color='g', label='Empleados por Tipo de Puesto')
ax[1].legend()

ax[0].set_title("Gráfico de Departamento")
ax[1].set_title("Gráfico de Puesto")

ax[0].set_xlabel("Tipo de departamento")
ax[0].set_ylabel("Cantidad de personas")

ax[1].set_xlabel("Tipo de puesto")
ax[1].set_ylabel("Cantidad de personas")

ax[0].tick_params(rotation = 45)
ax[1].tick_params(rotation = 45)

plt.tight_layout()

plt.savefig('graficas/graficas_departamento_puesto.jpg')
plt.show()
```

```python
fig,ax = plt.subplots(2,1, figsize = (20,30))


ax[0].plot(df_departamento["departamento"], df_departamento["count"], color = "black", marker = "^")

ax[1].plot(df_puesto["tipo_puesto"], df_puesto["count"], color ="black", marker = "^")
leg_mandos_medios, = ax[1].plot(df_puesto.iloc[0,0], df_puesto.iloc[0,1], color = "black", label = "mandos_medios")
leg_operativo, = ax[1].plot(df_puesto.iloc[1,0], df_puesto.iloc[1,1], color = "black", label = "operativo")
leg_administrativo, = ax[1].plot(df_puesto.iloc[2,0], df_puesto.iloc[2,1], color = "black", label = "administrativo")
leg_gerencia, = ax[1].plot(df_puesto.iloc[3,0], df_puesto.iloc[3,1], color = "black", label = "gerencia")


ax[0].set_title("Gráfico de Departamento")
ax[1].set_title("Gráfico de Puesto")

ax[0].set_xlabel("Tipo de departamento")
ax[0].set_ylabel("Cantidad de personas")

ax[0].tick_params(rotation = 45)
ax[1].tick_params(rotation = 45)

ax[1].set_xlabel("Tipo de puesto")
ax[1].set_ylabel("Cantidad de personas")


ax[1].legend(handles = [leg_mandos_medios, leg_operativo, leg_administrativo, leg_gerencia], labels = ['Mandos Medios', 'Operativo', 'Administrativo', 'Gerencia'])

plt.savefig('graficas/graficas_departamento_puesto2.jpg')
plt.show()
```
