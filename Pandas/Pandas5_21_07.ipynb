2. Chequead que los nombres de las columnas y ver si se llaman iguales entre dataframes. ¿Están todas en minúsculas? ¿Tienen espacios? En caso de que no sea así homogeneizar los nombres de las columnas para que todos los dataframes tengan el mismo formato.

```python
def cambio_columnas(dataframe):
    """
   Esta función cambia las columnas a minusculas sin espacios.
  Args: 
       dataframe(dataframe): el dataframe del cual queremos cambiar sus columnas.
   Returns: aunque no tiene return, esta función cambia las columnas a minusculas sin espacios reemplazando el anterior.
   """
    columnas_nuevas = {col : col.lower().replace(" ", "_") for col in dataframe.columns}
    dataframe.rename(columns = columnas_nuevas, inplace = True)
```

```python
cambio_columnas(df_2016)
cambio_columnas(df_2017)
cambio_columnas(df_2018)
cambio_columnas(df_2019)
cambio_columnas(df_2020)
cambio_columnas(df_agentes)
cambio_columnas(df_empleados)
cambio_columnas(df_puestos)
```
