2. Haced una exploración básica cada uno de los DataFrames, esta exploración debe incluir:
    * ¿Cuántas filas y columnas tiene el DataFrame?
    * Mostrad las primeras cinco filas y las últimas cinco filas del DataFrame.
    * Mostrad 2 filas al azar del DataFrame.
    * ¿Cuál es el nombre de las columnas del DataFrame?
    * ¿Cuáles son los principales estádisticos del DataFrame? Mostrad por separado las variables categóricas y las variables numéricas.
    * ¿Cuál es el porcentaje de valores nulos en cada columna?
    * ¿Cuántos valores únicos tenemos para cada una de las columnas?
    * De aquellas columnas que tengan menos de 6 elementos únicos mostrad cada uno de sus valores únicos.

```python
def propiedades(dataframe):
    """
   Esta función printea en pantalla todas las propiedades del dataframe.
  Args: 
       dataframe(dataframe): el dataframe del cual queremos saber sus propiedades.
   Returns: aunque no tiene return, esta printea en pantalla todas las propiedades del dataframe.
   """
    print("El número de filas y columnas es:", dataframe.shape)
    print(".......................................")
    print("Las primeras cinco filas son:")
    display(dataframe.head())
    print(".......................................")
    print("Las últimas cinco filas son:")
    display(dataframe.tail())
    print(".......................................")
    print("Dos filas al azar son:")
    display(dataframe.sample(2))
    print(".......................................")
    print("Los nombres de las columnas son:", dataframe.columns)
    print(".......................................")
    print("Los principios estadísticos son:")
    display(dataframe.describe(include = "object"))
    print(".......................................")
    print("El porcentaje de los valores nulos:", (dataframe.isnull().sum() / dataframe.shape[0]) * 100)
    print(".......................................")

    for columna in dataframe.columns:
        print("El número de valores únicos de la columna",columna, "es:", len(dataframe[columna].unique()))
        print("--------------------------------------------------")
        if len(dataframe[columna].unique()) < 6:
            print("Los valores únicos de la columna",columna, "son:", dataframe[columna].unique())
            print("**********************************")
```

```python
propiedades(tickets)
```

```python
propiedades(agentes)
```

```python
propiedades(empleados)
```

```python
propiedades(puestos)
```

3. Modificaciones básicas de los conjuntos de datos:
    * Cambiad el nombre de las columnas para que todas estén en minúsculas y reemplazad los espacios por _.

```python
def cambio_columnas(dataframe):
    """
   Esta función cambia las columnas a minusculas sin espacios.
  Args: 
       dataframe(dataframe): el dataframe del cual queremos cambiar sus columnas.
   Returns: aunque no tiene return, esta función cambia las columnas a minusculas sin espacios reemplazando el anterior.
   """
    columnas_nuevas = {col :  col.lower().replace(" ", "_") for col in dataframe.columns}
    dataframe.rename(columns = columnas_nuevas, inplace = True)
```

```python
cambio_columnas(tickets)
tickets.head()
```

```python
cambio_columnas(agentes)
agentes.head()
```

```python
cambio_columnas(puestos)
puestos.head()
```

```python
cambio_columnas(empleados)
empleados.head()
```
